```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# Load the dataset

df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")

# Display the first few rows of the dataset
print(df.head())

# Check for missing values
print(df.isnull().sum())
```

```python
# Convert categorical variables to appropriate data types
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
df['Contract'] = df['Contract'].astype('category')
```

```python
# Churn rate by tenure
sns.boxplot(x='Churn', y='tenure', data=df)
plt.title('Churn Rate by Tenure')
plt.show()
```

```python
# Churn rate by Internet service
sns.countplot(x='InternetService', hue='Churn', data=df)
plt.title('Churn Rate by Internet Service')
plt.show()
```

```python
# Churn rate by Contract
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title('Churn Rate by Contract')
plt.show()
```

```python
# Churn rate by Monthly Charges
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title('Churn Rate by Monthly Charges')
plt.show()
```

```python
churn_rate_per_tenure = df.groupby('tenure')['Churn'].mean() * 100

# Plotting the line chart
plt.plot(churn_rate_per_tenure.index, churn_rate_per_tenure, marker='o')

# Adding labels and title
plt.xlabel('Tenure (Months)')
plt.ylabel('Churn Rate (%)')
plt.title('Churn Rate Based on Tenure')

# Display the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

<h2>Observations</h2>
</br>
<p>Tenure Impact on Churn:</p>
<p>Customers with higher tenure are less likely to churn, as seen in the boxplot.</p>
</br>
<p>Internet Service Impact on Churn:</p>
<p>Fiber optic users have a higher churn rate compared to DSL or other services.</p>
</br>
<p>Contract Impact on Churn:</p>
<p>Month-to-month contracts have a higher churn rate compared to one or two-year contracts.</p>
</br>
<p>Monthly Charges Impact on Churn:</p>
<p>Customers with higher monthly charges are more likely to churn, as suggested by the boxplot.</p>
